# Objective
To scrape following data from the curator site (The Bay):
* product_title
* product_description
* product_brand
* product_images
* product_price
* product_material
* product_rating

# Importing Libraries

###  Ran the scraper and stored url in list

## Scraped products to gather the informations mentioned above

```python
option = webdriver.ChromeOptions()

# Run the argument with incognito
# option.add_argument(' â€” incognito')
driver = webdriver.Chrome(executable_path='chromedriver', chrome_options=option)


# Creating lists of features interested
product_title_list = list()
product_description_list = list()
product_brand_list = list()
product_image1_list =list()
product_image2_list =list()
product_image3_list =list()
product_price_list = list()
product_rating_list = list()


# Getting the start time to track on time required
start = time.time()
product_number = 0

driver.set_window_size(1000,900)
driver.get('https://www.thebay.com'+product_url_list[0])

# Wait 30 seconds for page to load and extract the element after it loads
    
timeout = 30
try:
    WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="consent-close"]')))
except TimeoutException:
    print('Timed out waiting for page to load')
    
    # ----------------------------click close------------------------------------
driver.find_element_by_xpath('//*[@id="consent-close"]').click()

# -------------------------------Web Scraping-------------------------------
for link in product_url_list: 
    # Open the url
    driver.set_window_size(1000,900)
    driver.get('https://www.thebay.com'+link)

    # Wait 30 seconds for page to load and extract the element after it loads
    
    timeout = 30
    try:
        WebDriverWait(driver, timeout).until(EC.visibility_of_element_located((By.XPATH,'/html/body/div[5]/div[2]/div[3]/div[3]/div[3]/div/div[2]/div/h1')))
    except TimeoutException:
        print('Timed out waiting for page to load')
        
    time.sleep(3)

    # -------------------------------Product title-------------------------------
    # find_elements_by_id returns an array of selenium objects.
    try:
        product_title = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[3]/div[3]/div[3]/div/div[2]/div/h1').text
    except:
        product_title = 'NA'
#     print("product title ",product_title)
    product_title_list.append(product_title)

    # -------------------------------Product price-------------------------------
    # This will return the product price
    try:
        product_price = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[3]/div[3]/div[3]/div/div[3]/div/div/div/span/span/span/span').text
    except:
        product_price = 'NA'
#     print("product price ",product_price)
    product_price_list.append(product_price)

    # -------------------------------Description---------------------------------
    # This will return the description of the product
    # click the "+" button to get description
    driver.find_element_by_class_name('svg-svg-59-plus-thin-dims.svg-svg-59-plus-thin.plus.pull-right').click()
    try:
        product_description = driver.find_element_by_id('collapsible-details-1').text
    except:
        product_description = 'NA'
#     print("product description ",product_description)
    product_description_list.append(product_description)
  
    # -------------------------------Brand---------------------------------------
    # This will return the brand of the product
    try:
        product_brand = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[3]/div[3]/div[3]/div/div[1]/div/h1/a').text
    except:
        product_brand = 'NA'
#     print("product brand ",product_brand)
    product_brand_list.append(product_brand)
    
    # -------------------------------Images---------------------------------------
    # This will return the brand of the product
    try:
        product_image1 = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[3]/div[3]/div[1]/div[2]/div/div/div[1]/img').get_attribute("src")
    except:
        product_image1 ='NA'
#     print("product image ",product_image1)
    product_image1_list.append(product_image1)
    
    try:
        product_image2 = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[3]/div[3]/div[1]/div[2]/div/div/div[2]/img').get_attribute("src")
    except:
        product_image2 ='NA'
#     print("product image ",product_image1)
    product_image2_list.append(product_image2)
    
    try:
        product_image3 = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[3]/div[3]/div[1]/div[2]/div/div/div[3]/img').get_attribute("src")
    except:
        product_image3 ='NA'
#     print("product image ",product_image1)
    product_image3_list.append(product_image3)
    
        # -------------------------------Rating---------------------------------------
    # This will return the rating of the product out of 5
    try:
        product_rating = driver.find_element_by_xpath('/html/body/div[5]/div[2]/div[3]/div[3]/div[5]/div/div[2]/div/div/div[2]/div[1]/div[1]/span[1]').text
    except:
        product_rating = 'NA'
#     print("product rating ",product_rating)
    product_rating_list.append(product_rating)
    
    time.sleep(5)
    product_number += 1
    print('The product number {} has been scraped successfully'.format(product_number))
    
    
    
# Let us make a panda dataframe of information scraped
data = {'link':product_url_list,'product_title': product_title_list,'price':product_price_list,'description':product_description_list,
       'brand':product_brand_list,'image1':product_image1_list,'image2':product_image2_list,'image3':product_image3_list,'rating':product_rating_list}
df_product = pd.DataFrame.from_dict(data)
df_product.index.name = 'id'
display(df_product)

# Generate time tracker print
end = time.time()
print("For {} links, the time taken is {}".format(len(product_url_list), end-start))

```

# Correcting the columns before exporting

## Correction of 'product link' column

```python
#correction of link
df_product['link'] = df_product['link'].apply(lambda x: 'https://www.thebay.com'+ x)

#correction of price of products on sale. took the before sale value as scraper didn't get the new price
df_product['price'] = df_product['price'].apply(lambda x: x.split()[3] if len(x)>10 else x)
```

## Parsing material information from description

```python
# function to pull out material information
def product_material(x):
    for line in x.splitlines():
        if '%' in line.partition('%'):
            x = line
            break
        else:
            x = 'NA'
    return x
```

```python
#created material column and extracted information from description
df_product['material'] = df_product['description'].apply(lambda x: product_material(x))
```
